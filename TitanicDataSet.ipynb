```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv("titanic.csv")
titanic
```

```python
titanic.head()
```

```python

titanic.isnull().sum()
```

```python

titanic.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)
```

```python
titanic.dropna(inplace=True)
```

```python
titanic.isnull().sum()
```

```python
titanic.shape
```

```python
titanic.describe()
```

```python
titanic.info()
```

```python
#Survived

plt.figure(figsize=(3,5))
sns.countplot(x='Survived',data=titanic,hue='Survived', width=0.4,palette='BuGn')
plt.grid(axis='y')
plt.title('Survived')
plt.show()
```

```python
#Gender

plt.figure(figsize=(3,5))
sns.countplot(x='Sex', data=titanic,hue='Sex',width=0.4,palette='RdPu')
plt.title('Gender')
plt.grid(axis='y')
plt.show()
```

```python
#Parch

plt.figure(figsize=(4,6))
sns.countplot(x='Parch',data=titanic,hue='Parch',palette='Paired')
plt.grid(axis='y')
plt.title('Parents / Children aboard the Titanic')
plt.show()
```

```python
#Pclass

plt.figure(figsize=(3,6))
sns.countplot(x='Pclass', data=titanic,width=0.6,hue='Pclass',palette='bright',saturation=0.5)
plt.grid(axis='y')
plt.title('Ticket Class')
plt.show()
```

```python
#SibSp

plt.figure(figsize=(4,6))
sns.countplot(x='SibSp',data=titanic,hue='SibSp')
plt.title('Siblings / spouses aboard the Titanic')
plt.grid(axis='y')
plt.show()
```

```python
#Age

titanic['Age'].value_counts()
```

```python
titanic['Age'].describe()
```

```python
plt.figure(figsize=(5,5))
sns.histplot(x='Age',data=titanic,color='b')
plt.axvline(37,color='r',label='Majority Of This Age People Travel')
plt.grid(axis='y')
plt.title('Age in years')
plt.legend()
plt.show()
```

```python
#Fare

titanic['Fare'].value_counts()
```

```python
titanic['Fare'].describe()
```

```python
plt.figure(figsize=(4,5))
plt.hist(titanic['Fare'],edgecolor='k')
plt.grid(axis='y')
plt.title('Passenger Fare')
plt.show()
```

```python
#Pclass vs Survived

data2 = titanic.groupby(['Pclass','Survived']).size().unstack()
data2
```

```python
data2.plot(kind='barh',width=0.4)
plt.grid(axis='x')
plt.show()
```

```python
#Sex vs Survived

data1= titanic.groupby(['Sex','Survived']).size().unstack()
data1
```

```python
data1.plot(kind='barh',width=0.3)
plt.ylabel('No. of people')
plt.xlabel('Gender')
plt.grid(axis='y')
plt.title('Sex vs Survived')
plt.show()
```

```python
titanic.head()
```

```python
#Multi-variance

sns.pairplot(titanic,vars=['Age','SibSp','Pclass'],hue='Sex',kind='reg')
plt.show()
```

```python
#Fare vs Survived

sns.kdeplot(titanic[titanic['Survived']==1]['Fare'],color='r',label='Survived')
sns.kdeplot(titanic[titanic['Survived']==0]['Fare'],color='c',label='Not Survived')
plt.legend()
plt.axvline(55,color='k')
plt.grid(axis='y')
plt.show()
```

```python
#Age vs Survived

sns.kdeplot(titanic[titanic['Survived']==1]['Age'],color='r',label='Survived')
sns.kdeplot(titanic[titanic['Survived']==0]['Age'],color='c',label='No Survived')
plt.axvline(33,color='k')
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
male_data=titanic[titanic['Sex']=='male']
female_data=titanic[titanic['Sex']=='female']
```

```python
male_data.head()
```

```python
female_data.head()
```

```python
sns.histplot(data=male_data,x='Age',color='c', label='Male', alpha=1)
sns.histplot(data=female_data,x='Age',color='r', label='Female', alpha=1)
plt.legend()
plt.grid(axis='y')
plt.show()
```
